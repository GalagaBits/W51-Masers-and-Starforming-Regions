```python
extracted_spectrum = cube[:,y,x]
```

```python
sp = pyspeckit.Spectrum(data=extracted_spectrum,
                        xarr=extracted_spectrum.spectral_axis)
```

```python
from astropy import units as u
```

```python
sp.plotter(xmin=22.9205*u.GHz, xmax=22.922*u.GHz)
```
