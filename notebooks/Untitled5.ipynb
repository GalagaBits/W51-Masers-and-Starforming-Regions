```python
#Imporing functions
import numpy as np
import matplotlib.pyplot as plt

#Creating sample datasets
x = [0,1,2,3,4,5] #List or np array
y = np.exp(x) 

x1 = np.linspace(0,5,100)
y1 = np.exp(x1) 

#Plotting Data
plt.style.use('dark_background')

fig, ax = plt.subplots()

ax.plot(x, y, linestyle='-',label='Secant line')
ax.plot(x1,y1,'limegreen', label='Function of growth {$\mathregular{e^{y}}$}')
ax.scatter(x,y,marker='x',color='red', s=140, label='Population sample')

#Labeling Axis in Matplotlib
fontsize = 14
ax.set_xlabel('Time (days)', fontsize=fontsize)
ax.set_ylabel('Population of Fungi (#)', fontsize=fontsize)
ax.set_title('Fungal Growth on Mars', fontsize=fontsize)

ax.grid(alpha=0.2, linestyle='-')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('https://fawn.ifas.ufl.edu/controller.php/week/obs/260;csv?asText=1')

time, temp, rel_hum, rain, flux = df.endTime, df.t2m, df.rh, df.rain, df.rfd

plt.style.use('dark_background')

plt.figure(figsize=(10, 6))

plt.suptitle("7-Day Weather History in Gainsville (FAWN)")

plt.subplot(2, 2, 1)
plt.title("AVG Temperature @ 2m")
plt.plot(time, temp, color='orange')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("AVG Relative Humidity @ 2m")
plt.plot(time, rel_hum, color='lightgreen')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Rainfall")
plt.plot(time, rain, color='lightblue')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("AVG Radiant Flux Density")
plt.plot(time, flux, color='yellow')
plt.axis('off')

print(df.endTime, df.t2m, df.rh, df.rain)
```
