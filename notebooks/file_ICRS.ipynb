```python
all_fits = glob.glob('*.pbcor.fits')

for fits_path in all_fits:
    with fits.open(fits_path) as hdul:
        hdr = hdul[0].header
        hdr['RADESYS'] = 'ICRS'

        base_name = os.path.splitext(fits_path)[0]  # Remove the .fits part
        new_fits_file = f"{base_name}_ICRS.fits"
    
        if 'EQUINOX' in hdr:
            del hdr['EQUINOX']

        hdul.writeto(new_fits_file, overwrite=True)

    print(f"Modified and saved: {new_fits_file}")
```

```python
import glob
import os
from astropy.io import fits

# Get all FITS files matching the pattern
all_fits = glob.glob('*.pbcor.fits')

for fits_path in all_fits:
    with fits.open(fits_path) as hdul:
        hdr = hdul[0].header
        data = hdul[0].data

        # Set the reference frame to ICRS
        hdr['RADESYS'] = 'ICRS'

        # Change the number of axes to 2 in the header
        hdr['NAXIS'] = 2

        # Remove frequency and extra axis keywords if present
        for key in ['NAXIS3', 'NAXIS4', 'CTYPE3', 'CTYPE4',
                    'CRVAL3', 'CRVAL4', 'CDELT3', 'CDELT4',
                    'CRPIX3', 'CRPIX4']:
            if key in hdr:
                del hdr[key]

        # Remove EQUINOX if it exists
        if 'EQUINOX' in hdr:
            del hdr['EQUINOX']

        # Option 1: If your data cube has singleton dimensions that you want to remove,
        # you can use squeeze() to collapse them:
        data2d = data.squeeze()
        if data2d.ndim != 2:
            # Option 2: If squeezing didn't yield 2D, then explicitly select the first slice
            # along the extra axes. Adjust indices if necessary.
            data2d = data[0, 0]
        
        # Update the data in the HDU
        hdul[0].data = data2d

        # Save to a new file
        base_name = os.path.splitext(fits_path)[0]  # Remove the .fits part
        new_fits_file = f"{base_name}_ICRS_X.fits"
        hdul.writeto(new_fits_file, overwrite=True)

    print(f"Modified and saved: {new_fits_file}")
```
