```python
import pyspeckit
from spectral_cube import SpectralCube
from astropy import units as u

import pylab as pl
pl.style.use('default')
```

## NH3(5,3) spw44

```python
cube = SpectralCube.read('W51North_spw_44_corrected2.image', format='casa_image')
cube = cube.with_spectral_unit(u.GHz)
y, x = 356, 638

sp = pyspeckit.Spectrum(xarr=cube.spectral_axis, data=cube[:, y, x])
fig = pl.figure(figsize=(12,8))
sp.plotter(color='black', xmin=21.281*u.GHz, xmax=21.282*u.GHz, figure=fig)

sp.specfit(guesses=[0.35, 21.28125e9, 0.00001e9, 0.025, 21.2814e9, 0.00001e9], linewidth=1.5)
sp.specfit(guesses=[0.35, 21.28125e9, 0.00001e9, 0.025, 21.2814e9, 0.00001e9], linewidth=1.5)

sp.specfit.plot_fit(lw=1.0, composite_lw=1.8, show_components=True)

pl.title("NH3 (5,3)_1 Line", fontsize=16)
pl.ylabel(f"$S_\\nu$ [Jy beam-1]",fontsize=16)
pl.xlabel(f"Frequency [GHz]",fontsize=16)
pl.xticks(fontsize=16)
pl.yticks(fontsize=16)

sp.specfit.plotresiduals(color='black', zeroline=True)
pl.ylim(-0.05, 0.05)
```
