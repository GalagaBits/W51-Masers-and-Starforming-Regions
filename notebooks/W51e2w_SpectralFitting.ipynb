```python
import pyspeckit
from spectral_cube import SpectralCube
from astropy import units as u
#Import data
cube = SpectralCube.read('/orange/adamginsburg/w51/vla/19A-254/derod/CASA/W51e2w_spw_8_rough_01.image.fits', format='fits')

import pylab as pl

cube = cube.with_spectral_unit(u.km / u.s, velocity_convention='radio', rest_value= 25.715141 * u.GHz)

#Pixel location (x,y)
```
