```python
def maser_points_axins():
    transform = axins.get_transform('world')
    
    # Read the CSV
    maser_points = pd.read_csv('w51_irs2_ammonia_maser_positions.csv')
    
    # Filter rows where 'region' == 'CA3'
    maser_points_CA3 = maser_points.loc[maser_points['region'] == 'CA3']
    
    # Iterate through filtered rows using iterrows()
    for _, row in maser_points_CA3.iterrows():
        axins.errorbar(
            row['x_mean'],
            row['y_mean'],
            xerr=row['x_mean_error'],
            yerr=row['y_mean_error'],
            color=row['color'],  # Use color from the 'color' column
            marker='+',
            transform=transform,
            label=row['line'], 
        )
    return
```

```python
def maser_points_axins2():
    transform = axins2.get_transform('world')
    
    # Read the CSV
    maser_points = pd.read_csv('w51_irs2_ammonia_maser_positions.csv')
    
    # Filter rows where 'region' == 'CA3'
    maser_points_W51North = maser_points.loc[maser_points['region'] == 'd2']
    
    # Iterate through filtered rows using iterrows()
    for _, row in maser_points_W51North.iterrows():
        axins2.errorbar(
            row['x_mean'],
            row['y_mean'],
            xerr=row['x_mean_error'],
            yerr=row['y_mean_error'],
            color=row['color'],  # Use color from the 'color' column
            marker='+',
            transform=transform,
            label = str(row['line'])
        )
    return
```
