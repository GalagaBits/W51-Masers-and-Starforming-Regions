## Step 4: Spectral Interpolation

We can choose either the spatial or spectral step first for interpolation.

First, match the resolution.

`np.diff()` calculates the difference between succeding vaules in a list in regards to a n-th order.

```python
velocity_res_1 = np.diff(cube1vel.spectral_axis)[0]
velocity_res_2 = np.diff(cube2vel.spectral_axis)[0]
velocity_res_1, velocity_res_2
```

Since the H2CS cube has "broader channels", we need to smooth the channel with of SiO before spatial reprojection. Find the appropriate FWHM.

```python
fwhm_gaussian = (velocity_res_1**2 - velocity_res_2**2)**0.5
fwhm_gaussian
```
