# SED d2

```python
import matplotlib.pylab as plt

# Band 6, Band 3, Q band
nu = [226, 86, 46.99]
amp = [2.106382375e-02 * 1000, 4.313024293e-02 * 1000, 5.348074374e-03 * 1000]
amp_error = [1.156574249e-03 * 1000, 9.608167991e-04  * 1000, 4.054327673e-04 * 1000 ]
```

```python
def SED_plot():
    fontsize = 16
    fig, ax = plt.subplots(figsize=(8,6))
    plt.title("d2 SED PBCOR", fontsize=fontsize)
    ax.tick_params(axis='both', which='major', labelsize=16)
    ax.set_xlabel("Frequency (GHz)", fontsize=fontsize)
    ax.set_ylabel("Flux (mJy/beam)", fontsize=fontsize)
    ax.errorbar(nu, amp, yerr=amp_error, fmt='x', color="black", capsize=3)
    plt.loglog()
    return ax
```

```python
# Power Law
import numpy as np

c1 = 0.5015144527
c2 = 5.831563403e-3
c3 = 6.780887678e-5

nu1 = np.linspace(45,250,100)
amp1 = c1*(nu1**1)
amp2 = c2*(nu1**2)
amp3 = c3*(nu1**3)
```

```python
ax = SED_plot()
plt.plot(nu1, amp1, label='$f =c\\nu$')
plt.plot(nu1, amp2, label='$f =c\\nu^2$', linestyle = '--')
plt.plot(nu1, amp3, label='$f =c\\nu^3$', linestyle = '--')
plt.loglog()

plt.legend()
```
