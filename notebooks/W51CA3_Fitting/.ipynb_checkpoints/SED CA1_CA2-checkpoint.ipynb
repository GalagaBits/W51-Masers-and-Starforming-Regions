# SED CA1

```python
import matplotlib.pylab as plt

# Band 6, Band 3, Q band
nu = [226, 86, 46.99]
amp = [5.736659023e-03 * 1000, 1.920761978e-03 * 1000, 2.373406750e-04 * 1000]
amp_error = [1.450548184e-04 * 1000,  3.035724229e-04 * 1000, 3.227781752e-05 * 1000 ]
```

```python
def SED_plot():
    fontsize = 16
    fig, ax = plt.subplots(figsize=(8,6))
    plt.title("CA1 SED", fontsize=fontsize)
    ax.tick_params(axis='both', which='major', labelsize=16)
    ax.set_xlabel("Frequency (GHz)", fontsize=fontsize)
    ax.set_ylabel("Flux (mJy/beam)", fontsize=fontsize)
    ax.errorbar(nu, amp, yerr=amp_error, fmt='x', color="black", capsize=3)
    plt.loglog()
    return ax
```

```python
# Power Law
import numpy as np

c1 = 0.0223344416
c2 = 2.597028094e-4
c3 = 3.019800108e-6

nu1 = np.linspace(45,250,100)
amp1 = c1*(nu1**1)
amp2 = c2*(nu1**2)
amp3 = c3*(nu1**3)
```

```python
ax = SED_plot()
plt.plot(nu1, amp1, label='$f =c\\nu$')
plt.plot(nu1, amp2, label='$f =c\\nu^2$', linestyle = '--')
plt.plot(nu1, amp3, label='$f =c\\nu^3$', linestyle = '--')
plt.loglog()

plt.ylim([0,15])
plt.legend()
```
