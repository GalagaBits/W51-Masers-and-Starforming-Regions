```python
from astropy.coordinates import SkyCoord
import astropy.units as u
import numpy as np
```

```python
sky_coord = SkyCoord(ra=list(df['x_mean'])*u.deg, dec=list(df['y_mean'])*u.deg, frame='icrs')
sky_coord_error = SkyCoord(ra=list(df['x_mean_error'])*u.deg, dec=list(df['y_mean_error'])*u.deg, frame='icrs')
```

```python
sky_coord.ra.to_string(unit=u.hour, sep=":", precision=4)
```

```python
sky_coord.dec.to_string(unit=u.degree, sep=":", precision=4)
```

```python
sky_coord_error.ra.to_string(unit=u.hour, sep=":", precision=4)
```

```python
sky_coord_error.dec.to_string(unit=u.degree, sep=":", precision=4)
```

```python
df['ra'] = sky_coord.ra.to_string(unit=u.hour, sep=":", precision=4)
df['dec'] = sky_coord.dec.to_string(unit=u.degree, sep=":", precision=4)
df['ra_error'] = sky_coord_error.ra.to_string(unit=u.hour, sep=":", precision=4)
df['dec_error'] = sky_coord_error.dec.to_string(unit=u.degree, sep=":", precision=4)
```

```python
df['lines'] = df['line']
```

```python
df['amp'] = df['amplitude']
```
