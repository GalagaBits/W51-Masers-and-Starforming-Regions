```python
import pyspeckit
from spectral_cube import SpectralCube
cube = SpectralCube.read('W51North_spw_28.image', format='casa_image')
import pylab as pl
pl.style.use('seaborn-white')
```

```python
pl.imshow(cube[:,480:530,480:530].max(axis=0).value)
```

```python
sp = pyspeckit.Spectrum(xarr=cube.spectral_axis, data=cube[:, 504, 484])
```

```python
sp.plotter(color='black')
```

```python
from astropy import units as u
```

```python
sp.plotter(color='black', xmin=25.7108*u.GHz, xmax=25.7115*u.GHz)
```

```python
fig = pl.figure(figsize=(12,8))
sp.plotter(color='black', xmin=25.7105*u.GHz, xmax=25.712*u.GHz, figure=fig)

sp.specfit(guesses=[0.1, 25.71092*1e9, 0.00001*1e9, 0.5, 25.7111e9, 0.00001e9, 0.2, 25.71125e9, 0.00001e9], linewidth=1.5)

sp.specfit.plot_fit(lw=1.0, composite_lw=1.5, show_components=True)
```

```python
fig = pl.figure(figsize=(12,8))
#sp.plotter(color='w', xmin=25.7105*u.GHz, xmax=25.712*u.GHz, figure=fig)
sp.specfit.plotresiduals(color='black')
pl.ylim(-0.05, 0.05)
```
