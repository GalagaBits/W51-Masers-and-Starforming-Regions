```python
import pyspeckit
from spectral_cube import SpectralCube
from astropy import units as u
#Import data
cube = SpectralCube.read('/orange/adamginsburg/w51/vla/19A-254/derod/CASA/W51e2w_spw_8_rough_01.image.fits', format='fits')

import pylab as pl

cube = cube.with_spectral_unit(u.GHz, velocity_convention='radio')
```

```python
fig = pl.figure(figsize=(12,8))

x, y = 326, 12

sp = pyspeckit.Spectrum(xarr=cube.spectral_axis, data=cube[:, y, x])
sp.plotter(color='black', figure=fig)
pl.show()
```

```python
from astropy import units as u

#Plot the gassiuan fits in to each peak in the spectra
fig = pl.figure(figsize=(12,8))
sp.plotter(color='black', xmin=23.6495*u.GHz, xmax=23.65005*u.GHz, figure=fig)

sp.specfit(guesses=[0.12, 23.64980, 0.00001],linewidth=1.5)
sp.specfit(guesses=[0.12, 23.64980, 0.00001],linewidth=1.5)

sp.specfit.plot_fit(lw=1.0, composite_lw=1.8, show_components=True)

pl.title("W51e2w Source #1", fontsize=16)
pl.ylabel(f"$S_\\nu$ [Jy beam-1]",fontsize=16)
pl.xlabel(f"Frequency [GHz]",fontsize=16)
pl.xticks(fontsize=16)
pl.yticks(fontsize=16)
#saveplotfig_gaussianfit()
```
