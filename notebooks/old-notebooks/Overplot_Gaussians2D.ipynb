```python
from matplotlib.transforms import (
    Bbox, TransformedBbox, blended_transform_factory)
from mpl_toolkits.axes_grid1.inset_locator import (
    BboxPatch, BboxConnector, BboxConnectorPatch)


def connect_bbox(bbox1, bbox2,
                 loc1a, loc2a, loc1b, loc2b,
                 prop_lines, prop_patches=None):
    if prop_patches is None:
        prop_patches = {
            **prop_lines,
            "alpha": prop_lines.get("alpha", 1) * 0.2,
            "clip_on": False,
        }

    c1 = BboxConnector(
        bbox1, bbox2, loc1=loc1a, loc2=loc2a, clip_on=False, **prop_lines)
    c2 = BboxConnector(
        bbox1, bbox2, loc1=loc1b, loc2=loc2b, clip_on=False, **prop_lines)

    bbox_patch1 = BboxPatch(bbox1, **prop_patches)
    bbox_patch2 = BboxPatch(bbox2, **prop_patches)

    p = BboxConnectorPatch(bbox1, bbox2,
                           # loc1a=3, loc2a=2, loc1b=4, loc2b=1,
                           loc1a=loc1a, loc2a=loc2a, loc1b=loc1b, loc2b=loc2b,
                           clip_on=False,
                           **prop_patches)

    return c1, c2, bbox_patch1, bbox_patch2, p


def zoom_effect01(ax1, ax2, xmin, xmax, **kwargs):
    """
    Connect *ax1* and *ax2*. The *xmin*-to-*xmax* range in both axes will
    be marked.

    Parameters
    ----------
    ax1
        The main axes.
    ax2
        The zoomed axes.
    xmin, xmax
        The limits of the colored area in both plot axes.
    **kwargs
        Arguments passed to the patch constructor.
    """

    bbox = Bbox.from_extents(xmin, 0, xmax, 1)

    mybbox1 = TransformedBbox(bbox, ax1.get_xaxis_transform())
    mybbox2 = TransformedBbox(bbox, ax2.get_xaxis_transform())

    prop_patches = {**kwargs, "ec": "none", "alpha": 0.2}

    c1, c2, bbox_patch1, bbox_patch2, p = connect_bbox(
        mybbox1, mybbox2,
        loc1a=3, loc2a=2, loc1b=4, loc2b=1,
        prop_lines=kwargs, prop_patches=prop_patches)

    ax1.add_patch(bbox_patch1)
    ax2.add_patch(bbox_patch2)
    ax2.add_patch(c1)
    ax2.add_patch(c2)
    ax2.add_patch(p)

    return c1, c2, bbox_patch1, bbox_patch2, p


def zoom_effect02(ax1, ax2, **kwargs):
    """
    ax1 : the main axes
    ax1 : the zoomed axes

    Similar to zoom_effect01.  The xmin & xmax will be taken from the
    ax1.viewLim.
    """

    tt = ax1.transScale + (ax1.transLimits + ax2.transAxes)
    trans = blended_transform_factory(ax2.transData, tt)

    mybbox1 = ax1.bbox
    mybbox2 = TransformedBbox(ax1.viewLim, trans)

    prop_patches = {**kwargs, "ec": "none", "alpha": 0.2}

    c1, c2, bbox_patch1, bbox_patch2, p = connect_bbox(
        mybbox1, mybbox2,
        loc1a=3, loc2a=2, loc1b=4, loc2b=1,
        prop_lines=kwargs, prop_patches=prop_patches)

    ax1.add_patch(bbox_patch1)
    ax2.add_patch(bbox_patch2)
    ax2.add_patch(c1)
    ax2.add_patch(c2)
    ax2.add_patch(p)

    return c1, c2, bbox_patch1, bbox_patch2, p


import matplotlib.pyplot as plt


fn = '/orange/adamginsburg/w51/2017.1.00293.S/may2021_imaging/w51n.spw0thru19.14500.robust0.thr0.075mJy.mfs.I.startmod.selfcal7.image.tt0'
img = SpectralCube.read(fn, format='casa_image')
cen = img.shape[1]/2, img.shape[2]/2
fig = pl.figure(figsize=(20,12))
fig, ax = quicklook(img[:,cen[0]-150:cen[0]+250, cen[1]-400:cen[1]+400], fig=fig, inner_stretch='linear')

ax.errorbar(290.9158956, 14.51800069, marker='', xerr=0.0000007744082441, yerr=0.0000007736500546, elinewidth=1, markersize=15, c='pink', transform=ax.get_transform('world'),label='(5,3)')
ax.errorbar(290.9159260, 14.51800888, xerr=0.000001654344601, yerr=0.000001653913398, marker='+',  markersize=15, color='pink' ,transform=ax.get_transform('world'))


ax.errorbar(290.9159003, 14.51801042,  xerr=0.000002231605111, yerr=0.000002351250058,marker='+',  markersize=15, color='orange', transform=ax.get_transform('world'),label='(6,2)')


ax.errorbar(290.9158939, 14.51800111,  xerr=0.0000007096939545, yerr=0.0000007087442309,marker='+',  markersize=15, color='white', transform=ax.get_transform('world'), label='(6,3)')
ax.errorbar(290.9159262, 14.51800959,  xerr=0.0000009249558173, yerr=0.000001004686655,marker='+',  markersize=15, color='white', transform=ax.get_transform('world'))

ax.errorbar(290.9158986, 14.51800816,  xerr=0.000001508011353, yerr=0.00000150789928,marker='+',  markersize=15, color='yellow', transform=ax.get_transform('world'),label='(7,4)')

ax.errorbar(290.9158982, 14.51801105,  xerr=0.0000007282855144, yerr=0.000000725438108,marker='+',  markersize=15, color='red', transform=ax.get_transform('world'),label='(7,5)')

ax.errorbar(290.9170759, 14.518231700,  xerr=0.0000009592657155, yerr=0.0000009525078288,marker='+',  markersize=15, color='blue', transform=ax.get_transform('world'),label='(7,6)')
ax.errorbar(290.9170768, 14.51823206,  xerr=0.000001154934128, yerr=0.000001147556912,marker='+',  markersize=15, color='blue', transform=ax.get_transform('world'))
ax.errorbar(290.9170765, 14.518231240,  xerr=0.000000935432323, yerr=0.0000009270761778,marker='+',  markersize=15, color='blue', transform=ax.get_transform('world'))

ax.errorbar(290.9170797, 14.51823187,  xerr=0.000002502666799, yerr=0.000002502605651,marker='+',  markersize=15, color='green', transform=ax.get_transform('world'),label='(7,7)')
ax.errorbar(290.9170811, 14.51823057,  xerr=0.000003297782113, yerr=0.000003297622733,marker='+',  markersize=15, color='green', transform=ax.get_transform('world'))
ax.errorbar(290.9170758, 14.51823158,  xerr=0.0000004803880607, yerr=0.0000004803625509,marker='+',  markersize=15, color='green', transform=ax.get_transform('world'))
ax.errorbar(290.9170760, 14.5182316,  xerr=0.0000007257750809, yerr=0.0000007257355513,marker='+',  markersize=15, color='green', transform=ax.get_transform('world'))

ax.errorbar(290.9158985, 14.51800926,  xerr=0.0000007257750809, yerr=0.000001031573272,marker='+',  markersize=15, color='cyan', transform=ax.get_transform('world'),label='(8,5)')

ax.errorbar(290.9158975, 14.51800925,  xerr=0.000001251703102, yerr=0.000001251358726,marker='+',  markersize=15, color='violet', transform=ax.get_transform('world'),label='(9,8)')


ax.errorbar(290.9170756, 14.51823135,  xerr=0.0000007157208018, yerr=0.0000007151353273,marker='+',  markersize=15, color='peachpuff', transform=ax.get_transform('world'),label='(11,9)')
ax.errorbar(290.916824, 14.51822272,  xerr=0.000001822484554, yerr=0.000001343610514,marker='+',  markersize=15, color='peachpuff', transform=ax.get_transform('world'))
ax.errorbar(290.9169891, 14.51815923,  xerr=0.000003270723414, yerr=0.000003270495024,marker='+',  markersize=15, color='peachpuff', transform=ax.get_transform('world'))



ax.legend()
ax2= ax.inset_axes([0.27, 0.65, 0.15, 0.30])
zoom_effect01(ax, ax2, 0, 1)

pl.show()

plt.show()
```
